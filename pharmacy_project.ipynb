### Отберу нужные колонки для продолжения работы

```python
df.columns
```

```python
my_columns = ['DR_Dat', 'DR_Tim', 'DR_NChk', 'DR_NDoc', 'DR_Apt', 'DR_NDrugs', 'DR_Kol', 'DR_CZak', 'DR_CRoz', 'DR_SDisc', 'DR_TPay', 'DR_CDrugs', 'DR_Suppl',
            'DR_CDisc', 'DR_BCDisc', 'DR_TabEmpl',
            'DR_VZak', 'DR_Pos']
```

```python
df = df[my_columns]
```

```python
df = df.rename(columns={'DR_Dat': 'dt',
                   'DR_Tim': 'c_time',
                   'DR_NChk': 'nchk',
                   'DR_NDoc': 'ndoc',
                   'DR_Apt': 'apt',
                   'DR_NDrugs': 'drug',
                   'DR_Kol': 'kol',
                   'DR_CZak': 'zak',
                   'DR_CRoz': 'roz',
                   'DR_SDisc': 'disc',
                   'DR_TPay': 'pay_type',
                   'DR_CDrugs': 'drug_id',
                   'DR_Suppl': 'suppl',
                   'DR_CDisc': 'disc_id',
                   'DR_BCDisc': 'disc_barcode',
                   'DR_TabEmpl': 'empl',
                   'DR_VZak': 'vzak',
                   'DR_Pos': 'pos'
})
```

```python
df.head()
```

### Я заметил, что типы некоторых колнок определились неправильно, нужно это исправить.

```python
df.dtypes
```

### Перевел колонку *disc_barcode*  , *disc_id* в тип *object*  и срезал ненужные символы, которые приписались библиотекой.

```python
df['disc_barcode'] = df['disc_barcode'].astype('str').replace('\.0', ' ', regex=True)
```

```python
df['disc_id'] = df['disc_id'].astype('str').replace('\.0', ' ', regex=True)
```

### Так же изменим колонку *dt*, переведем колонку в тип даты, при помощи функции *pd.to_datetime*.
### Далее применим функцию *strftime* для установки порядка записи даты.

```python
pd.to_datetime(df['dt']).dt.strftime('%d.%m.%Y')
```

### В столбце *vzak* храняться данные о покупке товаров *1: покупка в живую / 2: покупка в интернет-магаизне*

```python
df['vzak'].unique()
```

### Приведу значения данного столбца к более читаемому виду.

```python
df['vzak'].astype('str').replace('1', 'Обычный').replace('2', 'Интернет').unique()
```

```python
df['vzak'] = df['vzak'].astype('str').replace('1', 'Обычный').replace('2', 'Интернет')
```
